```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
import os
import pickle
import re
import math

%matplotlib inline
print('Library Loaded')
```

```python
def display(image):
    plt.imshow(image)
    plt.show()    
```

```python
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(image)
```

```python
pts1 = np.float32([[178,300],[10,400],[502,400],[334,300]])
pts2 = np.float32([[10,300],[10,400],[502,400],[502,300]])
M = cv2.getPerspectiveTransform(pts1,pts2)
topViewImg = cv2.warpPerspective(image,M,(512,512))
display(topViewImg)
```

```python
hsvImg = cv2.cvtColor(topViewImg,cv2.COLOR_RGB2HSV)
display(hsvImg)
lowerYellow = np.array([20,100,100],dtype="uint8")
upperYellow = np.array([30,255,255],dtype="uint8")
yellowMask = cv2.inRange(hsvImg,lowerYellow,upperYellow)
display(yellowMask)
```
